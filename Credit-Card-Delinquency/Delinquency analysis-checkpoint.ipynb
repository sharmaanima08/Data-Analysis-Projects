```python
def load_and_prepare_data(file_path):
    df = pd.read_csv(file_path)
    df['DATE'] = pd.to_datetime(df['DATE'])
    df['WEEK'] = df['DATE'].dt.to_period('W').dt.start_time
    weekly_summary = df.groupby(['STORE_NBR', 'WEEK']).agg(
        total_sales=('TOT_SALES', 'sum'),
        num_customers=('LYLTY_CARD_NBR', pd.Series.nunique)
    ).reset_index()

    return weekly_summary
```

```python
from scipy.stats import pearsonr
```

```python
def compare_trial_to_controls(weekly_summary, trial_store, metric='correlation', min_weeks=10):
    sales_pivot = weekly_summary.pivot(index='WEEK', columns='STORE_NBR', values='total_sales')
    cust_pivot = weekly_summary.pivot(index='WEEK', columns='STORE_NBR', values='num_customers')
    valid_weeks = sales_pivot[trial_store].dropna().index
    sales_pivot = sales_pivot.loc[valid_weeks]
    cust_pivot = cust_pivot.loc[valid_weeks]
    trial_sales = sales_pivot[trial_store]
    trial_customers = cust_pivot[trial_store]
    results = []
    
    for store in sales_pivot.columns:
        if store == trial_store:
            continue

        control_sales = sales_pivot[store]
        control_customers = cust_pivot[store]

        common_sales = trial_sales.notna() & control_sales.notna()
        common_customers = trial_customers.notna() & control_customers.notna()

        if common_sales.sum() < min_weeks or common_customers.sum() < min_weeks:
            continue

        if metric == 'correlation':
            sales_corr, _ = pearsonr(trial_sales[common_sales], control_sales[common_sales])
            cust_corr, _ = pearsonr(trial_customers[common_customers], control_customers[common_customers])
            score = (sales_corr + cust_corr) / 2
        elif metric == 'magnitude':
            sales_dist = np.mean(np.abs(trial_sales[common_sales] - control_sales[common_sales]))
            cust_dist = np.mean(np.abs(trial_customers[common_customers] - control_customers[common_customers]))
            score = (sales_dist + cust_dist) / 2
        else:
            raise ValueError("Metric must be 'correlation' or 'magnitude'.")

        results.append({'control_store': store, 'similarity_score': score})
    ascending = metric == 'magnitude'
    return pd.DataFrame(results).sort_values(by='similarity_score', ascending=ascending).reset_index(drop=True)
```

```python
if __name__ == "__main__":
    file_path = "QVI_data.csv"  
    weekly_data = load_and_prepare_data(file_path)
    trial_store = 77
    metric_type = 'correlation' 

    ranked_controls = compare_trial_to_controls(weekly_data, trial_store, metric=metric_type)
    print(f"Top control stores for trial store {trial_store} using {metric_type}:")
    print(ranked_controls.head(10))
```

```python
def select_control_store(weekly_summary, trial_store, pretrial_start, pretrial_end, top_n=5, plot=True):
    mask = (weekly_summary['WEEK'] >= pretrial_start) & (weekly_summary['WEEK'] <= pretrial_end)
    pretrial_data = weekly_summary[mask]
    sales_pivot = pretrial_data.pivot(index='WEEK', columns='STORE_NBR', values='total_sales')
    cust_pivot = pretrial_data.pivot(index='WEEK', columns='STORE_NBR', values='num_customers')
    trial_sales = sales_pivot[trial_store]
    trial_customers = cust_pivot[trial_store]
    results = []
    for store in sales_pivot.columns:
        if store == trial_store:
            continue

        control_sales = sales_pivot[store]
        control_customers = cust_pivot[store]

        common_sales = trial_sales.notna() & control_sales.notna()
        common_cust = trial_customers.notna() & control_customers.notna()

        if common_sales.sum() < 10 or common_cust.sum() < 10:
            continue
            
        sales_corr, _ = pearsonr(trial_sales[common_sales], control_sales[common_sales])
        cust_corr, _ = pearsonr(trial_customers[common_cust], control_customers[common_cust])
        mean_corr = (sales_corr + cust_corr) / 2

        sales_dist = np.mean(np.abs(trial_sales[common_sales] - control_sales[common_sales]))
        cust_dist = np.mean(np.abs(trial_customers[common_cust] - control_customers[common_cust]))
        mean_dist = (sales_dist + cust_dist) / 2
        score = mean_dist / mean_corr if mean_corr != 0 else np.inf

        results.append({
            'control_store': store,
            'sales_corr': sales_corr,
            'cust_corr': cust_corr,
            'mean_corr': mean_corr,
            'mean_dist': mean_dist,
            'score': score})

    ranked = pd.DataFrame(results).sort_values(by='score').reset_index(drop=True)

    if plot and not ranked.empty:
        top_controls = ranked.head(top_n)['control_store']
        for ctrl in top_controls:
            plt.figure(figsize=(12, 4))

            # Plot Sales
            plt.subplot(1, 2, 1)
            plt.plot(trial_sales.index, trial_sales.values, label=f'Trial Store {trial_store}')
            plt.plot(sales_pivot[ctrl].index, sales_pivot[ctrl].values, label=f'Control Store {ctrl}')
            plt.title('Total Sales Comparison')
            plt.xlabel('Week')
            plt.ylabel('Sales')
            plt.legend()

            # Plot Customers
            plt.subplot(1, 2, 2)
            plt.plot(trial_customers.index, trial_customers.values, label=f'Trial Store {trial_store}')
            plt.plot(cust_pivot[ctrl].index, cust_pivot[ctrl].values, label=f'Control Store {ctrl}')
            plt.title('Number of Customers Comparison')
            plt.xlabel('Week')
            plt.ylabel('Customers')
            plt.legend()

            plt.tight_layout()
            plt.show()

    return ranked.head(top_n)

weekly_data = load_and_prepare_data("QVI_data.csv")

# Define pre-trial period (assumed June 2018 to January 2019)
select_control_store(weekly_summary=weekly_data,trial_store=77,pretrial_start='2018-06-25',pretrial_end='2019-01-06',top_n=3)
```
